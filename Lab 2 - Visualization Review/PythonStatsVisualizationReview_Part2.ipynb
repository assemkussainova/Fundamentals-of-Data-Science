```python
# Read the house prices dataset into a dataframe
f = open('HousePrices.csv',newline='')
house = pd.read_csv(f)
house
```

```python
# Compute mode and median of 'SalePrice' using the corresponding formula.
# Compare the obtained results against pandas' built-in functions mode() and median()
def median(a):
    s = sorted(a)
    if len(s) % 2 == 1:
        return s[len(s)//2]
    else:
        return (s[len(s)//2]+s[len(s)//2-1])/2

print(median(house['SalePrice']))
print(house['SalePrice'].median())

def mode(a):
    a = list(a)
    return max(set(a), key = a.count)

print(mode(house['SalePrice']))
print(house['SalePrice'].mode())
```

```python
# Using matplotlib, plot a figure with two histograms of 'SalePrice' of all the houses with and without central air conditioning ('CentralAir')
# For both histograms, set bin size to 20 and opacity to 0.5
# Describe x-y labels, and place the legend on upper right side of the figure
withAir = house['SalePrice'][house.CentralAir == 'Y']
without = house['SalePrice'][house.CentralAir == 'N']
plt.hist([withAir, without], bins = 20, alpha = 0.5, label=['With central air', 'Without central air'])

plt.title('"SalePrice" of all the houses with and without central air conditioning')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# Using pandas, create a scatterplot of 'YearBuilt' (x) versus 'LotArea' (y) colored by 'SalePrice', and use a colormap named 'cool'
# Modify the size of points based on 'GarageArea'
house.plot.scatter(x = 'YearBuilt', y = 'LotArea', c = 'SalePrice', s = house['GarageArea'], cmap = 'cool')
```

```python
# Using matplotlib, plot the pie chart of roof styles 'RoofStyle'
# Among diferent roof styles explode the slice of 'Gable'
row_size = 7
column_size = 7
plt.figure(figsize = (column_size,row_size))

house_grouped = house.groupby('RoofStyle').size()
house_grouped
labels = house_grouped.index
sizes = house_grouped.values
explode = (0, 0.1, 0, 0, 0, 0)
plt.pie(sizes, startangle=90, autopct='%1.1f%%', explode=explode, pctdistance=1.2)
plt.title('RoofStyle of houses')
plt.legend(labels)
plt.show()
```

```python
# Using pandas, plot boxplot of 'SalePrice' for different roof styles 'RoofStyle'
# Rotate the x labels by 30 degrees, turn off the grids and set the fontsize to 10
house.boxplot(by = 'RoofStyle', column='SalePrice', rot=30, fontsize=10, grid=False)
plt.ylabel('')
plt.xlabel('')
plt.title('')
plt.show()
```

```python
# Now we consider another statistical function implemented in Pandas and commonly used for time series data
# Percetage change between the two values can be computed as: (current_value - previous_value)/previous_value

# Write a function that takes a Numpy matrix and returns the percetage change for each row 
# For each element in a row, treat it's prior element as a previous value.
# Since there are no elements prior to the first element in a row, set it's percetage change to np.Nan.
```

```python
def percent_change(mat):
    for j in range(0, 1):
        pc = np.nan
        print(pc)
    for j in range(1, len(mat)):
        pc = (mat[j]-mat[j-1])/mat[j-1]
        print(pc)
```

```python
# Make a line plot for the genderwise population of Kazakhstan, Kyrgyzstan and Uzbekistan. 
# Use 2 different line styles to represent genders, while keeping the color scheme consistent for each country
# Provide appropriate labels for each axis and each line in the legend
# Make sure the tick labels are set to years and are readible. (i.e. 1985 instead of 0)
row_size = 5
column_size = 10
fig = plt.figure(figsize = (column_size,row_size))

p = population.loc[population['Location'] == 'Kazakhstan']
x = np.array(p.iloc[0, 3:])
y = population.columns[3:]
plt.plot(y, x, linestyle = '--', color='green', label = 'Females in Kazakhstan')
plt.plot(y, np.array(p.iloc[1, 3:]), linestyle = '-', color='green', label = 'Males in Kazakhstan')

k = population.loc[population['Location'] == 'Kyrgyzstan']
xk = np.array(k.iloc[0, 3:])
plt.plot(y, xk, linestyle = '--', color='blue', label = 'Females in Kyrgyzstan')
plt.plot(y, np.array(k.iloc[1, 3:]), linestyle = '-', color='blue', label = 'Males in Kyrgyzstan')

u = population.loc[population['Location'] == 'Uzbekistan']
xu = np.array(u.iloc[0, 3:])
plt.plot(y, xu, linestyle = '--', color='red', label = 'Females in Uzbekistan')
plt.plot(y, np.array(u.iloc[1, 3:]), linestyle = '-', color='red', label = 'Males in Uzbekistan')

plt.xticks(rotation='vertical')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Make a line plot for the combined population of each country. 
# Provide appropriate labels for each axis and a legend that includes names of the countries
# Make sure the tick labels are set to years and are readible. (i.e. 1985 instead of 0)
row_size = 5
column_size = 10
fig = plt.figure(figsize = (column_size,row_size))

t = population.groupby('Location').sum()
x = np.array(t.iloc[0, 1:])
y = population.columns[3:]
plt.plot(y, x, linestyle = '-', label = 'Kazakhstan')
plt.plot(y, np.array(t.iloc[1, 1:]), linestyle = '-', label = 'Kyrgyzstan')
plt.plot(y, np.array(t.iloc[2, 1:]), linestyle = '-', label = 'Tajikistan')
plt.plot(y, np.array(t.iloc[3, 1:]), linestyle = '-', label = 'Turkmenistan')
plt.plot(y, np.array(t.iloc[4, 1:]), linestyle = '-', label = 'Uzbekistan')

plt.xticks(rotation='vertical')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
plt.plot(y, np.array(t.iloc[5, 1:]), linestyle = '-', label = 'World')

plt.xticks(rotation='vertical')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
# Make a line plot that shows percent change for all countries
# Add a dash line to the plot indicating the 0 change for all years
row_size = 5
column_size = 10
fig = plt.figure(figsize = (column_size, row_size))

p = population.groupby('Location').sum()
t = p.iloc[:, 1:].pct_change(axis=1)
x = np.array(t.iloc[0, :])
y = population.columns[3:]
plt.plot(y, x, linestyle = '-', label = 'Kazakhstan')
plt.plot(y, np.array(t.iloc[1, :]), linestyle = '-', label = 'Kyrgyzstan')
plt.plot(y, np.array(t.iloc[2, :]), linestyle = '-', label = 'Tajikistan')
plt.plot(y, np.array(t.iloc[3, :]), linestyle = '-', label = 'Turkmenistan')
plt.plot(y, np.array(t.iloc[4, :]), linestyle = '-', label = 'Uzbekistan')
plt.plot(y, np.array(t.iloc[5, :]), linestyle = '-', label = 'World')
plt.plot(y, np.zeros(35), linestyle = '--', label = '0 percentage change')

plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
